```python
import cv2
```

```python
cap = cv2.VideoCapture('6-3.mp4') # 입력할 영상 경로 + 파일명 입력
 
count = 1
 
while(cap.isOpened()):
    ret, image = cap.read()
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # grayscale변환
    image = cv2.flip(image, 1)
    
#     height, width = image.shape
#     matrix = cv2.getRotationMatrix2D((width/2, height/2), 90, 0.5)
#     image = cv2.warpAffine(image, matrix, (width, height))
    
    cv2.imwrite("6-3/frame%d.jpg" %count, image)
    print('Saved frame%d.jpg' %count)
    count += 1
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break 
        
cap.release()
cv2.destroyAllWindows()
```
